```python
import pandas as pd
```

```python
file_path = '/home/fractaluser/Fractal/data/airline/unzipped'
```

```python
df = pd.read_csv(file_path+'/1987.csv')
```

```python
df['date'] = df['DayofMonth'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Year'].astype(str)
```

```python
for c in df.date.unique():
    (df[df.date == c]).to_csv(file_path+'/'+c+'.csv', index=False)
```
